```python
def id_cleaner(df):
    id_list = df.id.unique()
    frames = []
    i = 0
    for eye in id_list:
        pdf = df[df.id == eye]
        pdf_left = pdf[pdf.Laterality == 'Left']
        pdf_right = pdf[pdf.Laterality == 'Right']
        pdf_left.id = i
        pdf_right.id = i + 1
        i += 2
        if len(pdf_left) != 0:
            frames.append(pdf_left)
        if len(pdf_right) != 0:
            frames.append(pdf_right)
    return pd.concat(frames)
```
